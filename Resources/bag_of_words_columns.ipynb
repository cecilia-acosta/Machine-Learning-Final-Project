```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "words": [
      ["aa", "ab", "ac"],
      ["aa", "ba", "bb"],
      ["ba", "bb", "bc"],
      ["ab", "ba", "bb"],
      ["ac", "ab", "ba"],
    ]
})
df
```

```python
df_2 = df.copy()
for index, row in df.iterrows():
    for word in row.words:
        if word not in df_2.columns:
            df_2[word] = 0
        df_2.loc[index, word] = 1
```

```python
df_2
```

```python
mylist = ['d','e','f']
```

```python
df_3 = pd.concat([df_2, pd.DataFrame(columns=mylist)])
df_3
```

```python
df_4 = df_3.copy()
for index, row in df_4.iterrows():
    for word in row.words:
        if word not in df_4.columns:
            df_4[word] = 0
        else:
            df_4[word] = 1
df_4
```
